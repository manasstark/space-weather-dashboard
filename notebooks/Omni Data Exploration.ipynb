```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
lst_file = "omni2_f8X1myCa23.lst"
fmt_file = "omni2_f8X1myCa23.fmt"
```

```python
with open(fmt_file, "r") as f:
    print(f.read())
```

```python
column_names = [
    "year",
    "doy",
    "hour",
    "bt",
    "bx_gsm",
    "by_gsm",
    "bz_gsm",
    "temperature",
    "density",
    "speed",
    "ey",
    "kp",
    "dst",
    "f107",
    "ae",
]
```

```python
df = pd.read_csv(
    lst_file,
    sep=r"\s+",
    header=None,
    names=column_names,
    engine="python"
)
```

```python
import pandas as pd

print(pd.__version__)
```

```python
missing_values = {
    "bt": 999.9,
    "bx_gsm": 999.9,
    "by_gsm": 999.9,
    "bz_gsm": 999.9,
    "temperature": 9999999,
    "density": 999.9,
    "speed": 9999,
    "ey": 999.99,
}

df = df.replace(missing_values, np.nan)
```

```python
df.isna().sum()
```

```python
df.describe().T
```

```python
df["datetime"] = (
    pd.to_datetime(df["year"], format="%Y")
    + pd.to_timedelta(df["doy"] - 1, unit="D")
    + pd.to_timedelta(df["hour"], unit="h")
)
```

```python
df[["year", "doy", "hour", "datetime"]].head()
```

```python
df.set_index("datetime", inplace=True)
```
